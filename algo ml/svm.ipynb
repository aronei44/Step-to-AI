SVM Gamma adalah parameter yang digunakan untuk mengontrol kehalusan keputusan. Semakin kecil nilai gamma, semakin halus keputusan. Semakin besar nilai gamma, semakin kasar keputusan. Semakin kasar keputusan, semakin banyak data yang dianggap sebagai noise. Sehingga, nilai gamma yang terlalu besar akan menyebabkan overfitting. Sebaliknya, nilai gamma yang terlalu kecil akan menyebabkan underfitting.
